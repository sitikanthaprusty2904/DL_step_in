```python

import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([par])
```

```python
len(tokenizer.word_index)
```

```python
seq=[]
for sen in par.split('\n'):
    dj=tokenizer.texts_to_sequences([sen])[0]
    for i in range(1,len(dj)):
        seq.append(dj[:i+1])
seq

```

```python
mx_ln=max([len(x) for x in seq])
mx_ln
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
pad_in=pad_sequences(seq,maxlen=mx_ln,padding='pre')
```

```python
X=pad_in[:,:-1]
y=pad_in[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=197)
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,LSTM
```

```python
model=Sequential()
model.add(Embedding(197,100))
model.add(LSTM(150))
model.add(Dense(197,activation='softmax'))
model.build(input_shape=(None, 32)) # Explicitly build the model imp
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
import numpy as np

text="Self control and discipline"
for i in range(7):
  text_=tokenizer.texts_to_sequences([text])[0]
  text__=pad_sequences([text_],maxlen=32,padding='pre')
  pos=np.argmax(model.predict(text__),axis=-1)
  word_=""
  for word,index in tokenizer.word_index.items():
    if index == pos:
      word_=word
      break

  text=text + " " + word_
  print(text)
```
